```python
def convert3(text):
    L = []
    # Check if input is already a list
    if isinstance(text, list):
        names = text
    else:
        # If it's a string, split it
        names = text.split(',')
    
    counter = 0
    for name in names:
        if counter < 3:
            # Remove all spaces within each name
            no_space_name = ''.join(name.strip().split())
            L.append(no_space_name)
        counter += 1
    return ','.join(L)  # Join with commas

# Apply the function
shows['cast'] = shows['cast'].apply(convert3)
shows
shows['tags']=shows['description']+shows['cast']+shows['country']+shows['director']+shows['listed_in']
new = shows.drop(columns=['description','cast','country','director','date_added','release_year','rating','duration','top3_cast','type','listed_in'])
new
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')

# Create vectors using your DataFrame directly
vector = tfidf.fit_transform(new['tags'])
similarity = cosine_similarity(vector)

def recommend(show):
    # First let's print the available titles containing our search term
    try:
        # Find the position in our current DataFrame
        show_position = new.reset_index()
        show_position = show_position[show_position['title'] == show].index[0]
        
        # Get distances
        distances = sorted(list(enumerate(similarity[show_position])), reverse=True, key=lambda x: x[1])
        
        print(f"\nTop 5 recommendations for '{show}':")
        for i in distances[1:6]:
            print(f"- {new.iloc[i[0]]['title']}")
            
    except IndexError:
        print(f"Show '{show}' not found")

# Test it
recommend('Valeria')
```
